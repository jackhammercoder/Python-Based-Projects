Calculate the number of funded companies and the number of total companies, sliced by industry and gender. To make it a bit easier to compare numbers for men and women, let's pivot gender from rows to columns, so that we can see the percentages for men and women side-by-side. Store this pivoted table in a variable; you'll need it for the next part.

```python
funded = data.groupby(['Industry', 'Entrepreneur Gender'])['Amount'].count().reset_index()
funded = funded.pivot_table(index = 'Industry', columns = 'Entrepreneur Gender')
funded
total = data.fillna(0).groupby(['Industry', 'Entrepreneur Gender'])['Amount'].count().reset_index()
total = total.pivot_table(index = 'Industry', columns = 'Entrepreneur Gender')
total
perc = funded / total
perc.fillna("None in group") ## the NaN's mean group had no people in it
```

Use your table to compute 95% confidence intervals for the difference between the percentage of funded companies for men and women, one interval for each industry. Are there any industries where the difference is especially stark and statistically significant?

**HINT**: If it's been a while since your last statistics course, you are looking for a confidence interval for the difference of two proportions. Make sure you used the "pooled" estimate in calculating the standard error.

```python
import math
from scipy.stats import norm
diff = perc.Amount.Male - perc.Amount.Female
z = norm.ppf(.975)
perc.fillna(0)
first = perc.Amount.Male * ( 1- perc.Amount.Male)
first = first / total.Amount.Male.count()
second  = perc.Amount.Female * (1 - perc.Amount.Female)
second = second / total.Amount.Female.count()
sesqr = first + second
se = sesqr**(.5)
lower = diff - (z * se)
upper = diff + (z * se)
perc['lower'] = lower
perc["upper"] = upper
perc
```

For Uncerain / other the interval is entirely above zero, thus we have statistically significant evidence that entreprenuers within this group that are males( i did males - females) have higer chance of being funded than females.
